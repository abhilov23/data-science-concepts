```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#DATA
```

```python
a = np.linspace(0,10, 11) #0 to 10 numbers
```

```python
b=a**4 #power of 4 at a
```

```python
fig = plt.figure()

#large set of AXES

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b) #here we created the plot on the matplotlib

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of four')



#Small set of AXES
axes2 = fig.add_axes([0.25, 0.25, 0.25, 0.25]) #this new set of axes will be placed between the bigger axes
axes2.plot(a, b) #plot for the smaller axes

axes2.set_xlim(0, 8)
axes2.set_ylim(0, 8000)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')



plt.show() #this hows overall plots created by the user
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st plot
plt.plot(x, y1, 'r')
plt.title("Sine Wave")

plt.subplot(2, 1, 2)  # 2nd plot
plt.plot(x, y2, 'b')
plt.title("Cosine Wave")

plt.tight_layout()  # Prevents overlap
plt.show()
```

```python
plt.plot(a, b)
plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 10)
plt.style.use('ggplot')  # Other options: 'seaborn', 'fivethirtyeight'

plt.show()
```

```python

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y1", color='g')
ax2.set_ylabel("Y2", color='b')

plt.show()

```

```python
plt.plot(x, y1)
plt.annotate('Max Point', 
             xy=(x[np.argmax(y1)], max(y1)), 
             xytext=(8, 0.5), 
             arrowprops=dict(facecolor='black'))
plt.show()
```

```python

plt.plot(x, y1, color='blue')
plt.fill_between(x, y1, color='blue', alpha=0.3)
plt.show()

```

```python

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()

```
