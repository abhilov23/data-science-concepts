```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('12_dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 4), dpi=200) #defining the plot using matplotlib, as it is already using matplotlib in the backend
#hue defined the different colors for the selected table in the dataframe
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2', size='salary', style='level of education')
#the palette can define the different color combinations you want to apply
#size can define the different size of the salary
```
