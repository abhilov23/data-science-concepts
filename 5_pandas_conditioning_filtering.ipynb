```python
def quality(total_bill, tip):  #creating a function to determine the quality of service based on total_bill and tip
    if tip / total_bill > 0.25:
        return 'good'
    elif tip / total_bill > 0.15:
        return 'ok'
    else:
        return 'bad'
```

```python
quality(16.99, 1.01)
```

```python
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1) # Applying the quality function to each row of total_bill and tip columns
```

```python
df['Quality']=df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1) # adding the quality column to the DataFrame
```

```python
df
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip']) # Using np.vectorize to apply the quality function to the entire DataFrame columns
#this is an alternative way to apply the function without using apply method, but this works faster for large datasets

df.head()
```

# Creating a code snippet for repeat calling

```python
import timeit
```

```python
# Code snipppet to be executed only once

setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('4_tips.csv')
def quality(total_bill, tip):
    if tip / total_bill > 0.25:
        return 'good'
    elif tip / total_bill > 0.15:
        return 'ok'
    else:
        return 'bad'
'''
```

```python
# code snippet whose execution time is to be measured

stmt_one = ''' 
df['Tip Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
'''

stmt_two = ''' 
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(setup=setup, stmt=stmt_one, number=1000) # Timing the execution of the first statement
```

```python
timeit.timeit(setup=setup, stmt=stmt_two, number=1000) # Timing the execution of the second statement
```
